## ***Uso de modelos lineales para tener en cuenta las variables correlacionadas con el pago***

Una vez que identificamos algunas variables que estn correlacionadas con la variable de objetivo, podemos usar un modelo lineal para capturar esta relacin cuantitativamente. Un modelo lineal encuentra la recta de mejor ajuste de los puntos dados.

```python
sns.lmplot(data = df, x="age_years", y="pay_yearly", ci=None, line_kws={"color": "blue"},
           scatter_kws={"color": "orange"},height=6, aspect=3)
plt.title("Pay vs. Age", fontsize=20, verticalalignment="bottom")
plt.xlabel("Age")
plt.ylabel("Pay")
plt.show()
```

Una recta tiene dos parmetros: intercepto ($\beta_0$) y pendiente ($\beta_1$), tambin conocidos como los **coeficientes** del modelo. Por lo tanto, un modelo lineal de salario frente a la edad se puede representar como:

$$ PAY{\_}YEARLY = \beta_0 + \beta_1 AGE{\_}YEARS + \varepsilon $$



## ***Interpretando la salida de un modelo lineal.***


La lnea recta que dibujamos en el diagrama de dispersin de pago vs. edad se ajust usando un modelo estadstico llamado **regresin lineal**, cuyos resultados se pueden ver a continuacin:

```python
## Modelo de regresin lineal simple.
formula1 = "pay_yearly ~ age_years"
model1 = smf.ols(formula = formula1, data = df).fit()
print(model1.summary())
```

En la significancia global examinamos la <br>
$H_0$ $\beta_1=0$ Es decir el modelo es slo la media. <br>
$H_1$ $\beta_1\neq0$ Es decir el modelo que incluye edad es significativo (aporta)

image.png

Se rechaza la hiptesis nula, el modelo es significativo.

La asimetra y la curtosis para los residuales estn cerca de los valores de una distribucin normal que sin respectivamente 0 y 3

image.png

La tabla de salida anterior contiene mucha informacin, solo debemos centrarnos en una pequea cantidad de cantidades. Estas son la variable de salida (tambin conocida como la **variable dependiente**), $R^2$ y los coeficientes (las estimaciones de $\beta_0$ y $\beta_1$) y sus $p$-valores.

### ***Coeficientes.***

El intercepto $\beta_0$ es de aproximadamente USD $\$63 840.$ Esto se puede considerar como el salario base; es decir, el salario esperado de una persona de edad cero. Con frecuencia, el intercepto no tiene una interpretacin significativa (como en este caso), eso est bien siempre que lo reconozcamos y tengamos una explicacin slida de por qu. La pendiente (el coeficiente $\beta_1$ para `age_years`) es USD $\$873,50$. La interpretacin de este coeficiente es:

***Si un empleado aumenta un ao su edad, se espera que su salario aumente en USD $\$ 873,50$ en promedio.***

Las variables de entrada como la `age_years` tambin se denominan **variables independientes** en el contexto de un modelo de regresin lineal.

### ***$R^{2}$***

Para el modelo de pago versus edad, $R^2 = 0.238$. Dado que este modelo solo explica alrededor del 23,8% de la variacin, esto nos motiva a investigar si se pueden utilizar otros factores adems de la edad para explicar las diferencias salariales y as mejorar nuestras predicciones.

Un hecho matemtico importante a tener en cuenta es que cuando tratamos con relaciones de dos variables, $R^2 \equiv \rho^2$, donde $\rho$ (rho) es el coeficiente de correlacin. En el diagrama de dispersin de salario frente a edad, el coeficiente de correlacin $\rho$ fue 0,4876 y, por lo tanto, $R^2=0,4876^2=0,238$. De esto se deduce que *cuanto ms fuerte sea la correlacin lineal entre dos variables, mejor ser la bondad de ajuste de la regresin lineal correspondiente*.

Hay muchas lneas posibles que podemos ajustar a nuestros datos, pero como dijimos antes, la que queremos es la lnea de mejor ajuste, que se define como la lnea (entre todas las lneas posibles) que maximiza el $R^2$. del modelo.

## ***En cuanto a la edad y el gnero.***

Ahora que hemos visto que la edad explica parte de la relacin con el salario, consideremos un modelo en el que tengamos en cuenta la edad y el gnero simultneamente. Pasaremos de ajustar una [**regresin lineal simple**](https://en.wikipedia.org/wiki/Simple_linear_regression) (una variable independiente, una variable dependiente) a ajustar una [**regresin lineal mltiple** ](https://www.investopedia.com/terms/m/mlr.asp) (varias variables independientes, una variable dependiente).

Las regresiones lineales pueden involucrar variables categricas tan fcilmente como las variables numricas. El truco consiste en codificar las categoras como nmeros para que el modelo pueda interpretarlas. Hay muchas formas de codificar variables categricas, pero una de los ms comunes es transformarlos en conjuntos de unos y ceros. Lo importante es tener en cuenta la forma en que interpretamos los coeficientes de las variables categricas en un modelo lineal que es ligeramente diferente a la de las variables numricas:

```python
temp = pd.get_dummies(df["male_female"], drop_first=True)
pd.concat([temp,df["male_female"]], axis = 1)
```

```python
## Modelo con solo la variable male_female
model_MF = smf.ols("pay_yearly ~ male_female", data = df)
fitted_MF = model_MF.fit()
print(fitted_MF.summary())
```

$$ PAY{\_}YEARLY = \beta_0 + \beta_1 AGE{\_}YEARS + \beta_2 {MALE{\_}FEMALE}\text{[T.M]} + \varepsilon $$


```python
## Modelo de regresin lineal mltiple.
formula2 = "pay_yearly ~ age_years + male_female"
model2 = smf.ols(formula = formula2, data = df).fit()
print(model2.summary())
```

La interpretacin del coeficiente de edad es similar a la anterior:

***Dentro de los grupos del mismo sexo, si la edad aumenta en un ao, se espera que el salario aumente USD $\$ 872$***

Observe que este valor es diferente del que obtuvimos en el anterior regresin: esto se debe a que la variabilidad presente en `pay_yearly` ahora tiene que dividirse entre ms variables explicativas.



Con respecto al coeficiente de la variable `male_female`. Muestra masculino (`T.M`) solamente, porque la categora femenina se toma como categora predeterminada (o referencia). Eso significa que el coeficiente representa el `pay_yearly` adicional o reducido que ocurre solo porque alguien es hombre en lugar de mujer. En ltima instancia, la eleccin de la categora predeterminada no importa: podramos haber optado fcilmente por hacer de masculino la categora predeterminada y, por lo tanto, el coeficiente de gnero sera `T.F`. El coeficiente $\$7398$ se interpreta de la siguiente manera:

***Para empleados de la MISMA EDAD, los hombres en promedio ganan \\$7,398 USD ms que las mujeres.***
